# Make the background plot for the poster

```python
import numpy as np
import gmt
```

I'll use the USGS quake data that comes with GMT.

```python
lon, lat, magnitude = np.loadtxt('usgs_quakes.txt', unpack=True)
```

Make a global Mercator map without any borders.

```python
fig = gmt.Figure()
fig.coast(region=[-270, 90, -70, 70], projection='M10i', land='#aaaaaa',
          water='white', resolution='l')
fig.plot(lon, lat, sizes=0.02*1.5**magnitude, style='cc', color=magnitude/magnitude.max(), cmap='ocean')
fig.show(width=900)
```

If it looks OK, save it to a high resolution PNG.

```python
fig.savefig('poster_background.png', dpi=1200)
```
